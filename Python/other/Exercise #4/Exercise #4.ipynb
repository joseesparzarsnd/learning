```python
import json
import csv
import glob

from typing import List
```

```python
def flatten_json(y: str) -> dict:
  out = {}

  def flatten(x: any, name: str = ''):
    if type(x) is dict:
      for a in x:
        flatten(x[a], name + a + '_')
    elif type(x) is list:
      i = 0
      for a in x:
        flatten(a, name + str(i) + '_')
        i += 1
    else:
      out[name[:-1]] = x
  flatten(y)
  return out
```

```python
def main() -> None:
  root_dir: str = f"{current_location}/data"
  json_paths: List[str] = glob.glob("**/*.json", root_dir = root_dir, recursive = True)
  for json_file in json_paths:
    file_path: str = json_file.replace("\\", "/")
    file_name: str = file_path.split("/")[-1]
    file_path = file_path.replace(file_name, "")
    complete_path: str = f"{root_dir}/{file_path}{file_name}"
    new_path: str = f"{root_dir}/{file_path}{file_name.replace('.json', '.csv')}"
    with open(complete_path, "r") as f, open(new_path, 'w', newline = '') as csv_file:
      data = json.load(f)
      data = flatten_json(data)
      columns = list(data.keys())
      csv_writer = csv.writer(csv_file)
      csv_writer.writerow(columns)
      csv_writer.writerow(list(data.values()))

if __name__ == "__main__":
  main()
```
