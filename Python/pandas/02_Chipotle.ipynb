```python
# Libraries.
import numpy as np
import pandas as pd
```

```python
# Import dataset from URL.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

# Assign data to a variable called chipo.
chipo = pd.read_csv(url, sep = '\t')

# Preview data.
chipo.head()
```

```python
# How many products cost more than $10.00?
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
print(f"A total of {(chipo.where(chipo['item_price'] > 10.00).count().iloc[0])} products cost more than $10.00")
```

```python
# What was the price of each item?
item_prices = chipo[['item_name', 'choice_description', 'item_price']] \
  .groupby(['item_name', 'choice_description']) \
  .max() \
  .drop_duplicates()

item_prices
```

```python
# What was the quantity of the most expensive item ordered?
most_expensive_item = item_prices.sort_values(by='item_price', ascending=False).head(1)

print(f"The most expensive item ordered was the {(most_expensive_item.index[0][0])}, which was ordered a total of {int(chipo.quantity.where(chipo['item_price'] == most_expensive_item['item_price'][0]).dropna().iloc[0])} times.")
```

```python
# How many times was a Veggie Salad Bowl ordered?
print("The Veggie Salad Bowl was ordered a total of", (chipo.where(chipo['item_name'] == "Veggie Salad Bowl").count()[0]) , "times.")

# How many times did someone order more than one canned soda?
print("Some ordered more than one canned soda a total of", chipo.where((chipo.item_name == "Canned Soda") & (chipo.quantity > 1)).dropna().count().iloc[0] ,"times.")
```
