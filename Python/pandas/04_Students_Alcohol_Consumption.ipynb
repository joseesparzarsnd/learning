```python
# Libraries.
import numpy as np
import pandas as pd
```

```python
# Import dataset from URL and assign it to a variable called df.
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv"

df = pd.read_csv(url)

# Preview data.
df.head()
```

```python
# Slice the dataframe from "school" until the "guardian" column.
columns_to_keep = []

for column in df.columns:
  columns_to_keep.append(column)
  if column.lower() == "guardian":
    break

columns_to_drop = set(df.columns).difference(set(columns_to_keep))

if set(df.columns).issuperset(columns_to_drop):
  df.drop(
      labels = columns_to_drop,
      axis = 1,
      inplace = True
  )

df.head()
```

```python
# Create a lambda function that will capitalize strings.
def capitalize(value):
  return str(value).upper()

# Capitalize both Mjob and Fjob.
df.Mjob.apply(lambda x: capitalize(x))
df.Fjob.apply(lambda x: capitalize(x))

# Print the last elements of the data set.
df.tail()
```

```python
# Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.
df.Mjob = df.Mjob.apply(lambda x: capitalize(x))
df.Fjob = df.Fjob.apply(lambda x: capitalize(x))

df.tail()
```

```python
# Create a function called majority that returns a boolean value to a new column called legal_drinker (consider majority as older than 17 years old).
def is_legal(age):
  return age > 17

df['legal_drinker'] = df.age.apply(lambda x: is_legal(x))

# Preview data.
df.head()
```

```python
# Multiply every number of the dataset by 10.
for column in df.columns:
  if(df[column].dtype == "int64"):
    df[column] = df[column].apply(lambda x: x * 10)

# Preview changes.
df.head()
```
