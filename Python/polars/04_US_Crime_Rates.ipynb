```python
# Import the dataset from the URL and assign it to a variable called crime.
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"

crime = pl.from_pandas(pd.read_csv(url))

# Preview data.
crime.limit(5)
```

```python
# What is the type of the columns?
crime.dtypes
```

```python
transformed = False
```

```python
# Convert the type of the column Year to datetime64.
if not transformed:
  crime = crime.with_columns(pl.col("Year").cast(pl.String).str.to_date(format = "%Y").cast(pl.Datetime))
  transformed = True

crime.limit(5)
```

```python
dropped = False
```

```python
# Delete the total column.
if not dropped:
  crime = crime.drop("Total")
  dropped = True

crime.limit(10)
```

```python
# Group the year by decades and sum the values.
# ** Note: Pay attention to the population column number, summing this column is a mistake.
crime \
  .with_columns(Decade = pl.col("Year").dt.year() - (pl.col("Year").dt.year() % 10)) \
  .group_by("Decade") \
  .agg(
      pl.col("Population").mean().round(0).cast(pl.Int64),
      pl.col(*[column for column in crimes.columns if column not in ["Population", "Year", "Decade"]]).sum()
  )
```

```python
# What's the most dangerous decade to live in the US?
crime \
  .with_columns(
      Decade = pl.col("Year").dt.year() - (pl.col("Year").dt.year() % 10),
      Total = pl.sum_horizontal(*[column for columns in crimes.columns if column not in ["Population", "Year", "Decade"]])
  ) \
  .group_by("Decade") \
  .agg(
      pl.col("Population").max(),
      pl.col(*[column for column in crimes.columns if column not in ["Population", "Year", "Decade"]]).sum(),
      pl.col("Total").sum()
  ) \
  .sort("Total", descending = True) \
  .select("Decade") \
  .limit(1)
```
