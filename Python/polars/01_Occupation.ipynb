```python
# Libraries.
import numpy as np
import polars as pl
```

```python
# Import data from URL.
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pl.read_csv(url, separator = "|")
```

```python
# See the first 25 entries.
users.head(25)
```

```python
# See the last 10 entries.
users.tail(10)
```

```python
# What is the number of observations in the dataset?
print(f"Number of observations: {users.shape[0]}")

# What is the number of columns in the dataset?
print(f"Number of columns: {users.shape[1]}")

# Print the name of all the columns.
print(f"Columns: {users.columns[0]}", *users.columns[1:], sep = ', ', end = ".\n")

# What's the data type for each column?
print("\nData types: ")
print(*[f" - {column}: {dtype}" for column, dtype in zip(users.columns, users.dtypes)], sep = "\n")
```

```python
# Print only the occupation column.
users.select("occupation").head(5)
```

```python
# How many different occupations are in this dataset?
print("Number of different occupations: ", users.select("occupation").n_unique())

# What is the most frequent occupation?
users \
  .select("occupation", "user_id") \
  .group_by("occupation").len(name = "count") \
  .sort("count", descending = True) \
  .head(1)
```

```python
# Summarize the dataframe.
users.describe()
```

```python
# Summarize only the occupation column.
users.select("occupation").describe()
```

```python
# What's the mean age of users?
users.select("age").mean()
```

```python
# What's the age with the least occurrence?
users \
  .select("age", "user_id") \
  .group_by("age").len(name = "count") \
  .sort("count", descending = False) \
  .head(1)
```
