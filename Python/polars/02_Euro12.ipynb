```python
# Import data from address.
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

# Assign data to a variable called euro12.
euro12 = pl.read_csv(url)

# Select only the Goal column.
euro12.select("Goals").head(10)
```

```python
# How many teams participated in the Euro2012?
print(f"A total of {euro12.select('Team').unique().shape[0]} participated in the Euro2012.")

# What is the number of columns in the dataset?
print(f"There are a total of {len(euro12.columns)} columns in the dataset.")
```

```python
# View only the columns Team, Yellow Cards and Red Cards, and assign them to a dataframe called discipline.
discipline = euro12.select(["Team", "Yellow Cards", "Red Cards"])
discipline.head(10)
```

```python
# Sort the teams by Red Cards, then to Yellow Cards.
discipline.sort("Red Cards").head(5)
```

```python
discipline.sort("Yellow Cards").head(5)
```

```python
# Calculate the mean Yellow Cards given per Team.
discipline.group_by("Team").mean().select("Team", "Yellow Cards")
```

```python
# Filter teams that scored more than 6 goals.
euro12.filter(pl.col("Goals") > 6).head()
```

```python
# Select the teams that start with a G.
discipline.filter(pl.col("Team").str.starts_with("G")).head()
```

```python
# Select the first 7 columns.
euro12.select(euro12.columns[:7]).head()
```

```python
# Select all columns except the last 3.
euro12.select(euro12.columns[:-3]).head()
```

```python
# Present only the Shooting Accuracy from England, Italy and Russia.
euro12.select("Team", "Shooting Accuracy").filter(pl.col("Team").is_in(["England", "Italy", "Russia"])).head()
```
